### <a id='toc1_1_1_'></a>[**Introduction**](#toc0_)
#### <a id='toc1_1_1_1_'></a>[**This Project**](#toc0_)
This project explores real-world data on oil and gas pipeline accidents in the United States, spanning from **2010 to 2023**. The dataset was sourced from **Kaggle** at [here](https://www.kaggle.com/datasets/energysafetystat/oil-facility-accidents-2010-present) and is based on actual reported incidents ‚Äî **not synthetic data**. It provides a comprehensive view of each accident, covering:

- üìç **Location & Geography**  
- üèóÔ∏è **Infrastructure & Technical Information**  
- ‚ö†Ô∏è **Incident Details** (e.g., causes, contributing factors)  
- ‚ò†Ô∏è **Human & Operational Impacts** (injuries, fatalities, disruptions)  
- üå± **Environmental Effects** (soil, water, air contamination)  
- üí∏ **Financial Consequences** (emergency response, property damage, cleanup costs)

---

#### <a id='toc1_1_1_2_'></a>[üß† **Project Goals**](#toc0_)

The aim of this project is to:
- **Clean** and prepare the dataset for analysis  
- **Explore and visualize** key trends, patterns, and risk factors  
- **Uncover insights** about causes, high-risk areas, and costly consequences  
- **Recommend strategies** to improve pipeline safety, reduce environmental impact, and manage financial risk

---

#### <a id='toc1_1_1_3_'></a>[üß∞ **Tools Used**](#toc0_)

This project was conducted using **Python**, leveraging:
- `NumPy` and `pandas` for data manipulation  
- `Matplotlib`, `Seaborn`, and `Plotly` for visual exploration and storytelling

---

By the end of the project, you'll find data-driven observations and actionable recommendations based on over a decade of pipeline accident data in the U.S.

**This initial exploration of the data to get a quick view of the data**

```python
df = pd.read_csv('accidents.csv')
df.columns = [col.lower() for col in df.columns]
data = df.copy()
```

```python
data.describe(include='all')
```

```python
cols_to_drop = []
for col in data.columns:
    if data[col].count() < 1500:
        cols_to_drop.append(col)

cols_to_drop.extend(["supplemental_number",'commodity_subtype','pipe_fac_name'])
```

## <a id='toc1_3_'></a>[**Analysis**](#toc0_)
### <a id='toc1_3_1_'></a>[**1. Occurence and Identification**](#toc0_)
#### <a id='toc1_3_1_1_'></a>[**Temporal Trends - Occurence**](#toc0_)

```python
data_dropped["occurence_date"] = data_dropped["local_datetime"].dt.date
weekly_occurence = data_dropped.groupby("occurence_date")[['report_number']].count().reset_index()
weekly_occurence["occurence_date"] = pd.to_datetime(weekly_occurence["occurence_date"])
fig, ax = plt.subplots(2,1, figsize=(15,12))
weekly_occurence.set_index("occurence_date").resample("ME").sum().plot(ax=ax[0], legend=False)
weekly_occurence.set_index("occurence_date").resample("ME").sum().rolling(12).mean().plot(ax=ax[0], legend=False)
ax[0].set_title("Monthly accidents - Total and rolling average")
ax[0].set_xlabel("")
weekly_occurence.set_index("occurence_date").resample("YE").sum().plot(ax=ax[1], legend=False)
ax[1].set_title("Yearly accidents")
ax[1].set_xlabel("");
```

#### <a id='toc1_3_1_2_'></a>[**Temporal Trends - Identification**](#toc0_)

```python
data_dropped['ident_mins'] = ((data_dropped["incident_identified_datetime"] - data_dropped["local_datetime"]).dt.total_seconds() / 60)
fig, ax = plt.subplots(1,2, figsize=(15,6))

data_dropped['ident_mins'].plot(kind='hist', ax=ax[0])
sns.violinplot(data=data_dropped, x='ident_mins', ax=ax[1]);
```

#### <a id='toc1_3_1_3_'></a>[**Insights**](#toc0_)

**Occurence:**
- From 2010 to 2016, incidents exhibited a general upward trajectory, peaking in 2016, followed by a gradual decline post-2016.  
- Despite the overall downward trend in yearly totals (with 2015 as the highest-incident year), short-term volatility is evident, including spikes in monthly occurrences (e.g., early 2021).
- A 12-month rolling average further corroborates the 2016 peak, highlighting a lag between short-term fluctuations and long-term trends.

**Identification:**
- Timely Identification: The majority of incidents are detected during or immediately after occurrence, reflecting robust real-time monitoring systems.
- Delayed Identification: A subset of incidents (e.g., leaks, corrosion failures) is identified weeks to months post-occurrence, suggesting gaps in routine inspections or sensor coverage.
- Preemptive Identification: A small fraction of incidents is flagged before escalation (e.g., via predictive maintenance or risk assessments), underscoring the potential of proactive mitigation strategies.

### <a id='toc1_3_2_'></a>[**2. Geography and Location**](#toc0_)

```python
location_count = data_dropped.groupby(["location_latitude","location_longitude"])[['report_number']].count().reset_index()

fig = px.scatter_geo(
    location_count,
    lat = 'location_latitude',
    lon = 'location_longitude',
    title = "Accidents by Location",
    projection = 'natural earth',
    color = 'report_number'
)

fig.update_geos(
    showland=True,               # Show land areas
    landcolor="lightgray",       # Land color
    showcountries=True,          # Show country borders
    countrycolor="black",        # Country border color
    showlakes=True,              # Show lakes
    lakecolor="blue",            # Lake color
    showocean=True,              # Show ocean
    oceancolor="lightblue",      # Ocean color
    showcoastlines=True,         # Show coastlines
    coastlinecolor="gray"        # Coastline color
)

fig.update_traces(marker=dict(
    colorscale='Reds',
    color=location_count['report_number'],
    showscale=True
))

# Manually set the color range (e.g., from 0 to the max value in 'Count')
fig.update_layout(
    coloraxis=dict(
        colorscale="Reds",  
        cmin=0,  # Minimum value
        cmax=5,  # Maximum value
        colorbar=dict(
            title="Count"
        )
    )
)

fig.show()
```

#### <a id='toc1_3_4_6_'></a>[**Insights**](#toc0_)
**Cause:**
- Equipment Failure is the most frequent cause overall, especially for non-carbon materials (blue bar dominates here).
- Corrosion Failure appears to be significantly more common in Carbon Steel (orange bar is much longer than the blue), highlighting its susceptibility to corrosion-related incidents.
- Incorrect Operation and Natural Force Damage similarly mostly associated with carbon materials as highlighted by the dominance of carbon.
- Material Failure of Pipe or Weld and Excavation Damage are minor contributors but are mostly associated with Carbon Steel (orange bars).
- Other Outside Force Damage and Other Accident Cause have low counts overall, but they are still also more frequent in Carbon Steel incidents.

If the distribution of accident causes was similar across both Carbon Steel and other materials, we could argue that the material type has little or no influence on the type of failure. However, in this analysis, the distributions differ significantly ‚Äî for example, Carbon Steel shows a higher rate of corrosion failures, while other materials experience more equipment-related issues. This suggests that the type of material can indeed influence the likely cause of an accident.

**Release types:**
- Leaks are the most frequent release type with carbon steel making up to more than half of the leak-related incidents.
- Other release types have very low counts compared to leaks with some of them having a high presence of the other material types

**Spills:**
- Crude oil spills are common with over 50% of all accidents, but about two-thirds of the spilled volume is recovered (~100,000 out of ~150,000 barrels). Still, the remaining ~50,000 barrels pose a serious environmental risk.
- Over 30% of the accidents involve refined liquids however, they have a lower recovery rate‚Äîonly about half is recovered (~35,000 out of ~70,000 barrels)‚Äîlikely because they evaporate or spread more easily.
- HVLs and toxic gases are released in large amounts (nearly 150,000 barrels), but recovery is minimal. Since these substances dissipate in the air or pose health risks, cleanup is difficult.
- CO‚ÇÇ and biofuel incidents are rare and involve small volumes, which may suggest safer handling or limited usage.

**4. Optimize Emergency Response Efficiency**
- Analyze high emergency response costs to identify optimization opportunities.
- Focus on improving coordination, resource allocation, and adopting advanced technologies.

**5. Material-Specific Safety Strategies**
- Develop tailored safety protocols and material selection guidelines.
- Address different failure modes: corrosion in carbon steel vs. equipment failure in other materials.

**6. Improve Environmental Remediation Practices**
- Review current soil remediation techniques to ensure effectiveness.
- Explore proactive measures to minimize soil contamination.

**7. Focus on High-Cost Incident Locations**
- Implement targeted prevention and enhanced safety protocols for:
  - Underground pipelines
  - Pipeline valve sites

**8. Data-Driven Prevention**
- Utilize detailed data on incident causes, locations, and consequences.
